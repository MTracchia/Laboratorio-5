```python

import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
filename = '/content/drive/My Drive/Repositorio de L5/Caminata aleatoria/bolita_individual.csv' 

t=[]
x=[]
y=[]

skip=0
with open(filename,"r") as csvfile:
  for row in csvfile:
    if skip >= 2:
      t.append(float(row.split(",")[0]))
      x.append(float(row.split(",")[1]))
      y.append(float(row.split(",")[2]))
    skip+=1

#X-Y
plt.figure(figsize=(10, 8),dpi=100)
plt.plot(x,y,marker='o')
#plt.tight_layout()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()

#X en funcion de t
plt.figure(figsize=(10, 8),dpi=100)
plt.plot(t,x,marker='o')
plt.grid()
plt.xlabel('Tiempo')
plt.ylabel('X')
plt._show()

#Y en funcion de t
plt.figure(figsize=(10, 8),dpi=100)
plt.plot(t,y,marker='o')
plt.xlabel('Tiempo')
plt.ylabel('Y')
plt.grid()
plt._show()

```
